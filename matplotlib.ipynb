```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,51)
print(x)
y=x
plt.plot(x,y,c="red",label = "y=(x)")
plt.legend()
plt.show()
```

```python
x = np.arange(1,51)
print(x)
y=(pow(x,2))
plt.plot(x,y,c="red",label = "y=(pow(x,2))")
plt.legend()
plt.show()
```

```python
y=4*pow(x,3)+x**2+4*x+7
plt.plot(x,y,c="red",label = "4*pow(x,3)+x**2+4*x+7")
plt.legend()
plt.show()
```

```python
y=1+(1/np.exp(-x))
plt.plot(x,y,c="red",label = "y=1+(1/np.exp(-x))")
plt.legend()
plt.show()
```

```python
y2=(pow(x,2))
y3=4*pow(x,3)+x+7
y4=1+(1/np.exp(x))
plt.plot(x,y2,c="red",label = "y=1+(1/np.exp(-x))")
plt.plot(x,y3,c="blue",label = "y=1+(1/np.exp(-x))")
plt.plot(x,y4,c="black",label = "y=1+(1/np.exp(-x))")
plt.legend()
plt.show()
```

```python
y=1+(1/np.exp(0))
plt.plot(x,y,c="red",label = "y=1+(1/np.exp(0))")
plt.legend()
plt.show()  
## x doesn't have any value we'll get error
```
